```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb, binom
from itertools import combinations
from scipy.optimize import curve_fit
from mpl_toolkits.axes_grid1 import ImageGrid
```

```python
n = 2 # Half-number of photons
N = 2*n
d = 2**N # Hilbert space dimension
C = comb(N,n)
S = ["V"]*2*n
L = np.arange(N)
print(S)
```

```python
comb_list = []

for c in combinations(L,n):
    s = np.copy(S)
    for i in c: s[i] = 'H'
    comb_list.append(s)
    
len(comb_list) == C
```

```python
H = np.array([1,0]); V = np.array([0,1])
L = np.array([1,1j])/np.sqrt(2); R = np.array([1,-1j])/np.sqrt(2)
get_vecA = {"H" : H, "V" : V}
get_vecB = {"H" : L, "V" : R}
rhoA = 0
rhoB = 0

for c in comb_list:
    print(*c)
    vec = 1
    
    [vec := np.outer(vec,get_vecA[l]).flatten() for l in c]
    rhoA += np.outer(vec,np.conjugate(vec))
    
    vec = 1
    
    [vec := np.outer(vec,get_vecB[l]).flatten() for l in c]
    rhoB += np.outer(vec,np.conjugate(vec))
    
rhoA = rhoA/C
rhoB = rhoB/C
```

```python
etaA = 0.5
etaB = 1-etaA

rho = etaB*rhoB - etaA*rhoA
plt.imshow(np.real(rho), cmap='bone')
plt.colorbar()
plt.show()
```

```python
eigvals, eigvecs = np.linalg.eig(rho)
eigvals = np.real(eigvals)

inds = eigvals.argsort()
eigvals = eigvals[inds[::-1]]
eigvecs = eigvecs[:,inds[::-1]]
```

```python
fig = plt.figure(figsize=(8,4))
color = "seismic"

grid = ImageGrid(fig, 111,          # as in plt.subplot(111)
                 nrows_ncols=(1,2),
                 axes_pad=0.15,
                 share_all=True,
                 cbar_location="right",
                 cbar_mode="single",
                 cbar_size="7%",
                 cbar_pad=0.15,
                 )

grid[0].imshow(np.real(eigvecs).T,cmap=color,vmin=-1,vmax=1)
grid[0].set_ylabel("Eigenvalues",fontsize=14)
grid[0].set_yticks(range(d),np.round(eigvals+1e-10,2),fontsize=12)
grid[0].set_xlabel("Basis vectors",fontsize=14)
grid[0].set_xticks([])
grid[0].set_title("Real part",fontsize=16)
im = grid[1].imshow(np.imag(eigvecs).T,cmap=color, vmin=-1,vmax=1)
grid[1].set_xlabel("Basis vectors",fontsize=14)
grid[1].set_xticks([])
grid[1].set_title("Imaginary part",fontsize=16)
#grid[1].set_yticks([])

# Colorbar
cbar = grid[1].cax.colorbar(im)
cbar.ax.tick_params(labelsize=12)
grid[1].cax.toggle_label(True)

plt.savefig('EigvecsMD.jpg', dpi=300)
#plt.tight_layout()    # Works, but may still require rect paramater to keep colorbar labels visible
plt.show()
```

## Conclusion:
La medida más óptima para discernir dos estados mixtos de 2n fotones es en general muy compleja y requiere de proyecciones sobre estados entrelazados multipartita. Es decir, requieren POVMs globales con forma compleja. En el caso de dos fotones, la POVM más óptima es una proyección sobre los estados de bell psi+-.

```python
def Perr_Despagnat(N): # N is the total
    return binom(N,N/2)/2**(N+1)
```

```python
Perr = [0.25,0.17783,0.14175,0.11966,0.10457,0.09352,0.08504]
nrange = 2*np.array(range(1,len(Perr)+1))
plt.figure()
plt.title("Minimum-error strategy", fontsize=16)
plt.scatter(nrange, Perr, marker='d', s=100, zorder=3, label="Optimal")
plt.scatter(nrange, Perr_Despagnat(nrange), marker='d', s=100, zorder=3, label="Optimal")
plt.ylabel(r"Error probability",fontsize=14)
plt.xlabel("Photon number",fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid()
plt.tight_layout()
plt.savefig('Perr_vs_N.jpg', transparent=True, dpi=300)
plt.show()
```

```python
def f(x, a, b, c):
    return a*x**b*np.exp(-c*x)

params, _ = curve_fit(f, nrange, Perr, (1,1,1))
```

```python
x = np.arange(2,1e3-200,2)
#plt.title("Minimum-error discrimination scheme", fontsize=16)
plt.plot(x, Perr_Despagnat(x), label="D'Espagnat's strategy")
plt.scatter(nrange, Perr, marker='d',color="tab:orange", zorder=3, label="Helstrom bound")
plt.plot(x, f(x,*params), color='k', ls='-.', alpha = 0.75, label="Helstrom bound (fit)")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel("Average error probability", fontsize=14)
plt.xlabel("Number of photons per box", fontsize=14)
plt.xscale('log')
plt.legend(fontsize=12)
plt.grid()
plt.savefig('Perr_vs_N.jpg', transparent=True, dpi=300)
plt.show()
```
