```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb
from itertools import combinations
from scipy.optimize import curve_fit
```

```python
n = 3 # Half-number of photons
N = 2*n
C = comb(N,n)
S = ["V"]*2*n
L = np.arange(N)
print(S)
```

```python
comb_list = []

for c in combinations(L,n):
    s = np.copy(S)
    for i in c: s[i] = 'H'
    comb_list.append(s)
    
len(comb_list) == C
```

```python
H = np.array([1,0]); V = np.array([0,1])
L = np.array([1,1j])/np.sqrt(2); R = np.array([1,-1j])/np.sqrt(2)
get_vecA = {"H" : H, "V" : V}
get_vecB = {"H" : L, "V" : R}
rhoA = 0
rhoB = 0

for c in comb_list:
    print(*c)
    vec = 1
    
    [vec := np.outer(vec,get_vecA[l]).flatten() for l in c]
    rhoA += np.outer(vec,np.conjugate(vec))
    
    vec = 1
    
    [vec := np.outer(vec,get_vecB[l]).flatten() for l in c]
    rhoB += np.outer(vec,np.conjugate(vec))
    
rhoA = rhoA/C
rhoB = rhoB/C
```

```python
etaA = 0.5
etaB = 1-etaA

rho = etaB*rhoB - etaA*rhoA
plt.imshow(np.real(rho), cmap='gray')
plt.colorbar()
plt.show()
```

```python
eigvals, eigvecs = np.linalg.eig(rho)
eigvals = np.real(eigvals)
```

```python
print(np.real(rho))
```

```python
#print(np.round(eigvals,10))
print("Perr = {:.5f}".format(0.5*(1 - np.sum(np.abs(eigvals)))))
```

```python
plt.scatter(range(2**N),eigvals)
plt.grid()
plt.show()
```

```python
plt.imshow(np.abs(eigvecs))
plt.colorbar()
plt.show()
```

## Conclusion:
La medida más óptima para discernir dos estados mixtos de 2n fotones es en general muy compleja y requiere de proyecciones sobre estados entrelazados multipartita. Es decir, requieren POVMs globales con forma compleja. En el caso de dos fotones, la POVM más óptima es una proyección sobre los estados de bell psi+-.

```python
Perr = [0.25,0.17783,0.14175,0.11966,0.10457,0.09352,0.08504]
nrange = range(1,len(Perr)+1)
plt.figure()
plt.title("Minimum-error strategy", fontsize=16)
plt.scatter(nrange, Perr, marker='d', s=100, zorder=3)
plt.ylabel(r"Error probability",fontsize=14)
plt.xlabel("N",fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid()
plt.tight_layout()
plt.savefig('Perr_vs_N.png', transparent=True)
plt.show()
```

```python
def f(x, a, b, c):
    return a*x**b*np.exp(-c*x)

params, _ = curve_fit(f, nrange, Perr, (1,1,1))
```

```python
x = np.arange(1,1e4,1)
plt.title("Minimum-error discrimination scheme", fontsize=16)
plt.scatter(nrange, Perr, marker='d', zorder=3)
plt.plot(x, f(x,*params), color='k', ls='-.', alpha = 0.5)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel("Optimal error probability", fontsize=14)
plt.xlabel("Photon number", fontsize=14)
plt.xscale('log')
plt.grid()
plt.show()
```
