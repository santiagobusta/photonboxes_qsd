```python
from sympy import *
from sympy.physics.quantum import TensorProduct
```

```python
var("s1 s2 s3 s4 s5 s6")
var("a1 a2 a3 a4", real=True)

# lower case for 2x2 matrices, upper case for 6x6 matrices

alpha = pi/8 # half angle between the states
opt_a2 = asin(tan(alpha))/2

S0 = Matrix([s1,s2,s3,s4,s5,s6]) # initial state vector
idd = eye(2)
IDD = eye(6)
zero = idd*0
qwp1 = Matrix([[1,0],[0,-I]])
hwp1 = Matrix([[cos(2*a1),sin(2*a1)],[sin(2*a1),-cos(2*a1)]]).subs(a1,3*alpha/2)
hwp2 = Matrix([[cos(2*a2),sin(2*a2)],[sin(2*a2),-cos(2*a2)]])#.subs(a2,opt_a2)
hwp3 = Matrix([[cos(2*a3),sin(2*a3)],[sin(2*a3),-cos(2*a3)]]).subs(a3,pi/4)
hwp4 = Matrix([[cos(2*a4),sin(2*a4)],[sin(2*a4),-cos(2*a4)]]).subs(a4,3*pi/8)
```

```python
# wave plates
QWP1 = Matrix(BlockMatrix([[qwp1,zero,zero],[zero,idd,zero],[zero,zero,idd]]))
HWP1 = Matrix(BlockMatrix([[hwp1,zero,zero],[zero,idd,zero],[zero,zero,idd]]))
HWP2 = Matrix(BlockMatrix([[hwp2,zero,zero],[zero,idd,zero],[zero,zero,idd]]))
HWP3 = Matrix(BlockMatrix([[idd,zero,zero],[zero,hwp3,zero],[zero,zero,idd]]))
HWP4 = Matrix(BlockMatrix([[idd,zero,zero],[zero,hwp4,zero],[zero,zero,idd]]))

# polarizing beam splitters
PBS1 = IDD.copy()
PBS1[1,1] = PBS1[3,3] = 0
PBS1[1,3] = PBS1[3,1] = 1

PBS2 = IDD.copy()
PBS2[1,1] = 0
PBS2[5,5] = 0
PBS2[1,5] = 1
PBS2[5,1] = 1

PBS3 = IDD.copy()
PBS3[3,3] = PBS3[5,5] = 0
PBS3[3,5] = PBS3[5,3] = 1
```

```python
# Preparation stage (QWP1 and HWP1)
 
S1 = HWP1.multiply(QWP1.multiply(S0))
S1
```

```python
# Stage 2 (PBS1 and HWP2 and HWP3)

S2 = HWP3.multiply(HWP2.multiply(PBS1.multiply(S1)))
S2
```

```python
# Stage 3 (PBS2 and PBS3 and HWP4)

S3 = HWP4.multiply(PBS3.multiply(PBS2.multiply(S2)))
S3
```

# Introducing a photon in the second port

```python
var("theta", real=True)

S_test = S3.subs([(s3,cos(theta)),(s4,sin(theta)),(s5,0),(s6,0)]) # no photons in the third port, a vertical photon in the second port
S_test
```

```python
SH = S_test.subs([(s1,1),(s2,0)]) # Horizontal photon
simplify(SH)
```

```python
SV = S_test.subs([(s1,0),(s2,1)]) # Vertical photon
simplify(SV)
```

```python
SL = S_test.subs([(s1,1/sqrt(2)),(s2,I/sqrt(2))]) # Left circular photon
simplify(SL)
```

```python
SR = S_test.subs([(s1,1/sqrt(2)),(s2,-I/sqrt(2))]) # Right circular photon
simplify(SR)
```

```python
PH = SH.multiply_elementwise(conjugate(SH))
PV = SV.multiply_elementwise(conjugate(SV))
PA = simplify(PH+PV)/2
PA
```

```python
PL = SL.multiply_elementwise(conjugate(SL))
PR = SR.multiply_elementwise(conjugate(SR))
PB = simplify(PL+PR)/2
PB
```

```python
simplify((PA-PB))
```

```python
# Fidelity between the two
# Minimizing the fidelity allows for a better discrimination
F = 0
for i in range(6):
    F += sqrt(PA[i]*PB[i])
```

```python
plotting.plot3d(F,(theta,0,2*pi),(a2,0,2*pi))
```

# Conclusiones

Sabemos que un fotn V en el segundo puerto de entrada permite diferenciar las tres cajas. Al parecer, un fotn A en el puerto 2 junto con un HWP1 en un ngulo 3alfa/2 en vez de alfa/2 podran permitir una mejor distincin. Tambin un fotn L en el puerto 2 con un HWP1 en su valor por default (alfa/2).

Valores por default:

a1 = alpha/2

a2 = opt_a2

a3 = pi/4

a4 = 3pi/4

theta = pi/2


```python
plotting.plot(F.subs(a2,opt_a2)-1,(theta,0,2*pi))
```

# Conclusin

Un mejor parmetro para la distincin es usar HWP1 en un ngulo 3alfa/2 con el fotn V en el puerto 2. A pesar de que la fidelidad disminuye para otros ngulos, este ngulo permite que no haya que usar un detector extra en la salida del puerto 3.
